```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from matplotlib import dates as mdates


# Load the data with explicit date format
dataset_train = pd.read_csv('MSFT.csv', parse_dates=['Date'], dayfirst=True)

# Verify the data
print(dataset_train.head())


# Extracting Date and Adj Close columns
training_set = dataset_train[['Date', 'Close']].copy()
training_set['Date'] = pd.to_datetime(training_set['Date'])
training_set = training_set.set_index('Date')

# Filtering data for the time period between 2018 and 2023
selected_data = training_set.loc['2018-01-01':'2023-12-31']

# Converting dates to numeric values
selected_data['NumericDate'] = mdates.date2num(selected_data.index)

# Creating the Linear Regression model
model = LinearRegression()
model.fit(selected_data[['NumericDate']], selected_data['Close'])

# Forecasting for the next year within the selected time period
future_dates = pd.date_range(start=selected_data.index[-1] + pd.Timedelta(days=1), periods=365, freq='D')
future_numeric_dates = mdates.date2num(future_dates)
forecasted_prices = model.predict(future_numeric_dates.reshape(-1, 1))

# Plotting
plt.figure(figsize=(12, 8))

# Historical Data
plt.plot(selected_data.index, selected_data['Close'], label='Historical Data', color='blue', marker='o')

# Forecasted Data
plt.plot(future_dates, forecasted_prices, label='Forecasted Data', color='red', linestyle='--')

# Scatter plot for individual data points
plt.scatter(selected_data.index, selected_data['Close'], color='blue', marker='o', label='Data Points')

plt.xlabel('Date', fontsize=12)
plt.ylabel('Microsoft Stock Price', fontsize=12)
plt.legend()
plt.xticks(rotation=45, ha='right')

plt.grid(True)
plt.title('Historical Data and Forecasted Values (2018-2023)', fontsize=14)
plt.show()


# Detailed Evaluation
print("Historical Data (2022-2023):")
print(selected_data.tail())  # Print the last few rows of historical data
print("\nForecasted Values (2024):")
print(pd.DataFrame({'Date': future_dates, 'Forecasted Price': forecasted_prices}).set_index('Date').tail())  # Print the first few rows of forecasted values


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV file
file_path = 'eurusd.csv'  # Replace with the actual path to your CSV file
df = pd.read_csv(file_path)

# Convert 'Date' to datetime format
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

# Sort DataFrame by 'Date' for better visualization
df = df.sort_values(by='Date')

# Plotting
plt.figure(figsize=(12, 8))
plt.plot(df['Date'], df['Close'], marker='o', linestyle='-', color='blue')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Close', fontsize=12)
plt.title('euro to usd rate forecasted for 2024', fontsize=14)
plt.grid(True)
plt.show()
print("detailed evaluation of last values:")
print(df.tail())
```
